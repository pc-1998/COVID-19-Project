```python
confirmed_last_day=time_series_confirmed.loc[:,['FIPS','4/18/20']]
deaths_last_day=time_series_deaths.loc[:,['FIPS','4/18/20']]
abridged_couties_clean=abridged_couties.drop(index=[3242,3243]) #City1, NY and City2, NY, both with no data
abridged_couties_clean['countyFIPS']=abridged_couties_clean['countyFIPS'].astype('int')
confirmed_last_day_clean=confirmed_last_day.copy()
confirmed_last_day_clean['FIPS']=confirmed_last_day_clean['FIPS'].replace(to_replace=np.nan,value=0)
confirmed_last_day_clean['FIPS']=confirmed_last_day_clean['FIPS'].astype('int')
wanted_data=abridged_couties_clean.merge(confirmed_last_day_clean,how='inner',left_on='countyFIPS',right_on='FIPS')
deaths_last_day_clean=deaths_last_day.copy()
deaths_last_day_clean['FIPS']=deaths_last_day_clean['FIPS'].replace(to_replace=np.nan,value=0)
deaths_last_day_clean['FIPS']=deaths_last_day_clean['FIPS'].astype('int')
wanted_data=wanted_data.merge(deaths_last_day_clean,how='inner',left_on='countyFIPS',right_on='FIPS')
wanted_data=wanted_data.drop(columns=['countyFIPS','COUNTYFP','CountyName',\
                                      'State','FIPS_y','FIPS_x'])
wanted_data=wanted_data.rename(columns={'4/18/20_x':'confirmed_4/18','4/18/20_y':'deaths_4/18'})
wanted_data['STATEFP']=wanted_data['STATEFP'].astype('int')
apr18states_wanted=apr18states.loc[:,['FIPS','Testing_Rate','Hospitalization_Rate', 'Incident_Rate']]
apr18states_wanted['FIPS']=apr18states_wanted['FIPS'].replace(to_replace=np.nan,value=0)
wanted_data=wanted_data.merge(apr18states_wanted,how='left',left_on='STATEFP',right_on='FIPS')
wanted_data=wanted_data.drop(columns=['FIPS','STATEFP'])
wanted_data=wanted_data.drop(wanted_data.loc[wanted_data['StateName']=='AK'].index)
wanted_data=wanted_data.drop(wanted_data.loc[wanted_data['StateName']=='HI'].index)
wanted_data=wanted_data.drop(wanted_data.loc[wanted_data['confirmed_4/18']<10].index)
wanted_data
```

```python
wanted_clean = wanted_data.copy()
wanted_clean = wanted_clean.loc[:,['stay at home', 'Incident_Rate', 'Testing_Rate', 'Hospitalization_Rate', 'PopulationEstimate2018', 'PopulationDensityperSqMile2010', '#Hospitals', '#ICU_beds', 'confirmed_4/18', 'deaths_4/18']]
wanted_clean['Hospital_per_capita'] = wanted_clean.apply(lambda row: row['#Hospitals'] / row['PopulationEstimate2018'], axis=1)
wanted_clean['ICU_bed_per_capita'] = wanted_clean.apply(lambda row: row['#ICU_beds'] / row['PopulationEstimate2018'], axis=1)
wanted_clean.drop(['#Hospitals', '#ICU_beds'], axis=1, inplace=True)
wanted_clean.fillna(wanted_clean.mean(), inplace=True)
wanted_clean['stay at home'] = wanted_clean['stay at home'].apply(lambda x: x - np.mean(wanted_clean['stay at home']))
wanted_clean['death_rate'] = wanted_clean.apply(lambda row: row['deaths_4/18'] / row['confirmed_4/18'], axis=1)
wanted_clean['death_rate'].fillna(0, inplace=True)
states=pd.get_dummies(wanted_data['StateName'])
states.loc[:,:'WV']
for c in states.loc[:,:'WV'].columns:
    wanted_clean[c]=states[c]
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
from math import sqrt

train, test = train_test_split(wanted_clean, test_size=0.1, random_state=42)
X_train = train.loc[:,['Testing_Rate', 'Incident_Rate', 'Hospitalization_Rate', 'stay at home', 'PopulationEstimate2018', 'PopulationDensityperSqMile2010', 'Hospital_per_capita', 'ICU_bed_per_capita']]
#X_train = train.drop(columns=['deaths_4/18','death_rate'])
Y_train = train.loc[:,['death_rate']]

model =  LinearRegression()
model.fit(X_train, Y_train)
Y_pred = model.predict(X_train)
Y_pred = np.clip(Y_pred, a_min=0, a_max=1)
RMSE = sqrt(mean_squared_error(Y_train, Y_pred))
print("Training RMSE Loss: ", RMSE)
```

```python
# Compare the RMSE loss to the range of the standardized Y_train
sd = np.std(np.array(Y_train))
mu = np.mean(np.array(Y_train))
std_Y_train = (Y_train - mu) / sd
std_range = max(std_Y_train['death_rate']) - min(std_Y_train['death_rate'])
std_range
```
