```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style = "whitegrid", 
        color_codes = True,
        font_scale = 1.5)
%matplotlib inline

apr18states = pd.read_csv('final_proj_data/covid19/4.18states.csv')
abridged_couties = pd.read_csv('final_proj_data/covid19/abridged_couties.csv')
time_series_confirmed = pd.read_csv('final_proj_data/covid19/time_series_covid19_confirmed_US.csv')
time_series_deaths = pd.read_csv('final_proj_data/covid19/time_series_covid19_deaths_US.csv')
```

```python
#11: starts the dates
from matplotlib.pyplot import figure
figure(num=None, figsize=(55, 20), dpi=80, facecolor='w', edgecolor='k')
plt.plot(np.sum(time_series_confirmed[time_series_confirmed['Country_Region']=='US'].iloc[:,51:], axis=0))
plt.title('cumulative confirmed cases in the US');
```

```python
#12 starts the dates
figure(num=None, figsize=(55, 20), dpi=80, facecolor='w', edgecolor='k')
plt.plot(np.sum(time_series_deaths[time_series_confirmed['Country_Region']=='US'].iloc[:,52:],axis=0))
plt.title('cumulative deaths in the US');
```

```python
#time_series_confirmed[time_series_confirmed['Province_State']=='New York']
#12: starts the dates
figure(num=None, figsize=(55, 20), dpi=80, facecolor='w', edgecolor='k')
plt.plot(np.sum(time_series_confirmed[time_series_confirmed['Province_State']=='California'].iloc[:,52:],axis=0))
plt.xticks(rotation=90)
plt.title('CA confirmed cumulative')
plt.show()
figure(num=None, figsize=(55, 20), dpi=80, facecolor='w', edgecolor='k')
plt.plot(np.sum(time_series_deaths[time_series_deaths['Province_State']=='California'].iloc[:,52:],axis=0))
plt.xticks(rotation=90)
plt.title('CA deaths cumulative')
plt.show();
figure(num=None, figsize=(55, 20), dpi=80, facecolor='w', edgecolor='k')
plt.plot(np.sum(time_series_confirmed[time_series_confirmed['Province_State']=='New York'].iloc[:,52:],axis=0))
plt.xticks(rotation=90)
plt.title('NY confirmed cumulative')
plt.show()
figure(num=None, figsize=(55, 20), dpi=80, facecolor='w', edgecolor='k')
plt.plot(np.sum(time_series_deaths[time_series_deaths['Province_State']=='New York'].iloc[:,52:],axis=0))
plt.title('NY deaths cumulative')
plt.xticks(rotation=90)
plt.show()
figure(num=None, figsize=(55, 20), dpi=80, facecolor='w', edgecolor='k')
plt.plot(np.sum(time_series_confirmed[time_series_confirmed['Province_State']=='Louisiana'].iloc[:,52:],axis=0))
plt.xticks(rotation=90)
plt.title('LA confirmed cumulative')
plt.show()
figure(num=None, figsize=(55, 20), dpi=80, facecolor='w', edgecolor='k')
plt.plot(np.sum(time_series_deaths[time_series_deaths['Province_State']=='Louisiana'].iloc[:,52:],axis=0))
plt.xticks(rotation=90)
plt.title('LA deaths cumulative')
plt.show()
```

```python
confirmed_last_day=time_series_confirmed.loc[:,['FIPS','4/18/20']]
deaths_last_day=time_series_deaths.loc[:,['FIPS','4/18/20']]
abridged_couties_clean=abridged_couties.drop(index=[3242,3243])
abridged_couties_clean[abridged_couties_clean['countyFIPS']=='City2']
abridged_couties_clean['countyFIPS']=abridged_couties_clean['countyFIPS'].astype('int')
confirmed_last_day_clean=confirmed_last_day.copy()
confirmed_last_day_clean['FIPS']=confirmed_last_day_clean['FIPS'].replace(to_replace=np.nan,value=0)
confirmed_last_day_clean['FIPS']=confirmed_last_day_clean['FIPS'].astype('int')
wanted_data=abridged_couties_clean.merge(confirmed_last_day_clean,how='inner',left_on='countyFIPS',right_on='FIPS')
deaths_last_day_clean=deaths_last_day.copy()
deaths_last_day_clean['FIPS']=deaths_last_day_clean['FIPS'].replace(to_replace=np.nan,value=0)
deaths_last_day_clean['FIPS']=deaths_last_day_clean['FIPS'].astype('int')
wanted_data=wanted_data.merge(deaths_last_day_clean,how='inner',left_on='countyFIPS',right_on='FIPS')
wanted_data=wanted_data.drop(columns=['countyFIPS','COUNTYFP','CountyName',\
                                      'State','FIPS_y','FIPS_x'])
wanted_data=wanted_data.rename(columns={'4/18/20_x':'confirmed_4/18','4/18/20_y':'deaths_4/18'})
wanted_data['STATEFP']=wanted_data['STATEFP'].astype('int')
apr18states_wanted=apr18states.loc[:,['FIPS','Testing_Rate','Hospitalization_Rate']]
apr18states_wanted['FIPS']=apr18states_wanted['FIPS'].replace(to_replace=np.nan,value=0)
wanted_data=wanted_data.merge(apr18states_wanted,how='left',left_on='STATEFP',right_on='FIPS')
wanted_data=wanted_data.drop(columns=['FIPS','STATEFP'])
wanted_data
```

```python
pd.options.display.max_columns = None
pd.options.display.max_rows = 20
wanted_data
```

```python
banLargeImpact = wanted_data[['>500 gatherings', 'confirmed_4/18', 'deaths_4/18']]

sns.set(style="whitegrid")

f, ax = plt.subplots(figsize=(6.5, 6.5))
sns.despine(f, left=True, bottom=True)

sns.scatterplot(x=">500 gatherings", y="confirmed_4/18",
                hue="deaths_4/18",
                palette="ch:r=-.2,d=.3_r",
                sizes=(1, 5), linewidth=0,
                data=banLargeImpact, ax=ax)
plt.title('When >500 gatherings were banned');
```

```python
medicalImpact = wanted_data[['MedicareEnrollment,AgedTot2017','PopulationEstimate65+2017', 'confirmed_4/18', 'deaths_4/18']]
medicalImpact['medRatio'] = medicalImpact['MedicareEnrollment,AgedTot2017'] / medicalImpact['PopulationEstimate65+2017'] 

sns.set(style="whitegrid")

f, ax = plt.subplots(figsize=(10, 5))
sns.despine(f, left=True, bottom=True)

sns.scatterplot(x="medRatio", y="confirmed_4/18",
                hue="deaths_4/18", size = 'MedicareEnrollment,AgedTot2017',
                palette="ch:r=-.2,d=.3_r",
                sizes=(1, 5), linewidth=0,
                data=medicalImpact, ax=ax)
plt.title('Effect of medical');
```
